# Deal with missing data
- Either we can drop row and column containing NAN values or it can be replace with mean or frequency
   - Drop
   - Replace
   
 **Best way is to replace missing data with mean**
 
-  normalized-losses    41   ==> Mean
-  bore                  4   ==> Mean
-  stroke                4   ==> Mean
-  horsepower            2   ==> Mean
-  peak-rpm              2   ==> Mean
-  price                 4   ==> delete ==> 
-  num-of-doors          2   ==> Mean

```python
df["normalized-losses"].replace(np.nan,df["normalized-losses"].astype("float").mean(),inplace = True)
df["bore"].replace(np.nan,df["bore"].astype("float").mean(),inplace = True)
df["stroke"].replace(np.nan,df["stroke"].astype("float").mean(),inplace = True)
df["horsepower"].replace(np.nan,df["horsepower"].astype("float").mean(),inplace = True)
df["peak-rpm"].replace(np.nan,df["peak-rpm"].astype("float").mean(),inplace = True)
```

Some columns are not of the correct data type. 
- Numerical variables should have type 'float' or 'int'.
- variables with strings such as categories should have type 'object'. 

For example, 'bore' and 'stroke' variables are numerical values that describe the engines, It can be type of 'float' or 'int'; but its showing type as 'object'. 

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```
